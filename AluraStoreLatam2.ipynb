### Importación de datos

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación

```python
ingresos_tienda1 = (tienda['Precio'] - tienda['Costo de envío']).sum()
ingresos_tienda2 = (tienda2['Precio'] - tienda2['Costo de envío']).sum()
ingresos_tienda3 = (tienda3['Precio'] - tienda3['Costo de envío']).sum()
ingresos_tienda4 = (tienda4['Precio'] - tienda4['Costo de envío']).sum()

print("Ingresos Totales:")
print(f"Tienda 1: ${ingresos_tienda1:,.2f}")
print(f"Tienda 2: ${ingresos_tienda2:,.2f}")
print(f"Tienda 3: ${ingresos_tienda3:,.2f}")
print(f"Tienda 4: ${ingresos_tienda4:,.2f}")
```

# 2. Ventas por categoría

```python
print("Categorías más vendidas:")
print("Tienda 1:")
print(tienda['Categoría del Producto'].value_counts().head(3))
print("\nTienda 2:")
print(tienda2['Categoría del Producto'].value_counts().head(3))
print("\nTienda 3:")
print(tienda3['Categoría del Producto'].value_counts().head(3))
print("\nTienda 4:")
print(tienda4['Categoría del Producto'].value_counts().head(3))
```

# 3. Calificación promedio de la tienda

```python
print("Calificación promedio de los clientes:")
print(f"Tienda 1: {tienda['Calificación'].mean():.2f}")
print(f"Tienda 2: {tienda2['Calificación'].mean():.2f}")
print(f"Tienda 3: {tienda3['Calificación'].mean():.2f}")
print(f"Tienda 4: {tienda4['Calificación'].mean():.2f}")
```

# 4. Productos más y menos vendidos

```python
print("Productos más vendidos:")
print("Tienda 1:")
print(tienda['Producto'].value_counts().head(3))
print("\nTienda 2:")
print(tienda2['Producto'].value_counts().head(3))
print("\nTienda 3:")
print(tienda3['Producto'].value_counts().head(3))
print("\nTienda 4:")
print(tienda4['Producto'].value_counts().head(3))

print("\nProductos menos vendidos:")
print("Tienda 1:")
print(tienda['Producto'].value_counts().tail(3))
print("\nTienda 2:")
print(tienda2['Producto'].value_counts().tail(3))
print("\nTienda 3:")
print(tienda3['Producto'].value_counts().tail(3))
print("\nTienda 4:")
print(tienda4['Producto'].value_counts().tail(3))
```

# 5. Envío promedio por tienda

```python
print("Costo de envío promedio:")
print(f"Tienda 1: ${tienda['Costo de envío'].mean():,.2f}")
print(f"Tienda 2: ${tienda2['Costo de envío'].mean():,.2f}")
print(f"Tienda 3: ${tienda3['Costo de envío'].mean():,.2f}")
print(f"Tienda 4: ${tienda4['Costo de envío'].mean():,.2f}")
```

```python
import matplotlib.pyplot as plt

# 1. Gráfico de barras para los ingresos totales
ingresos = [ingresos_tienda1, ingresos_tienda2, ingresos_tienda3, ingresos_tienda4]
tiendas_nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(10, 6))
plt.bar(tiendas_nombres, ingresos)
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos (en millones)')
plt.show()

# 2. Gráficos circulares para las ventas por categoría
fig, axs = plt.subplots(2, 2, figsize=(15, 15))

axs[0, 0].pie(tienda['Categoría del Producto'].value_counts(), labels=tienda['Categoría del Producto'].value_counts().index, autopct='%1.1f%%')
axs[0, 0].set_title('Tienda 1: Ventas por Categoría')

axs[0, 1].pie(tienda2['Categoría del Producto'].value_counts(), labels=tienda2['Categoría del Producto'].value_counts().index, autopct='%1.1f%%')
axs[0, 1].set_title('Tienda 2: Ventas por Categoría')

axs[1, 0].pie(tienda3['Categoría del Producto'].value_counts(), labels=tienda3['Categoría del Producto'].value_counts().index, autopct='%1.1f%%')
axs[1, 0].set_title('Tienda 3: Ventas por Categoría')

axs[1, 1].pie(tienda4['Categoría del Producto'].value_counts(), labels=tienda4['Categoría del Producto'].value_counts().index, autopct='%1.1f%%')
axs[1, 1].set_title('Tienda 4: Ventas por Categoría')

plt.show()

# 3. Gráfico de barras para la calificación promedio de los clientes
calificaciones = [tienda['Calificación'].mean(), tienda2['Calificación'].mean(), tienda3['Calificación'].mean(), tienda4['Calificación'].mean()]

plt.figure(figsize=(10, 6))
plt.scatter(tiendas_nombres, calificaciones)
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Calificación Promedio')
plt.show()
```
